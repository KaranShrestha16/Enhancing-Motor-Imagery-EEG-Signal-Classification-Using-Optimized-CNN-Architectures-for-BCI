```python
import mne
import numpy as np
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, concatenate
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
import seaborn as sns
import matplotlib.pyplot as plt

# Data Loading and Preprocessing
base_path = r'C:\\Users\\karan\\Downloads\\EEG Data\\Data'
subjects = [f'A0{i}' for i in range(1, 10) if i != 4]
event_ids = [7, 8, 9, 10]  # Define event IDs for motor imagery tasks

all_features = []
all_labels = []

for subject in subjects:
    file_path = f'{base_path}\\{subject}T.gdf'
    print(f"Processing {subject}...")
    raw = mne.io.read_raw_gdf(file_path, preload=True)
    raw.drop_channels(['EOG-left', 'EOG-central', 'EOG-right'])
    raw.set_eeg_reference()
    events, _ = mne.events_from_annotations(raw)
    epochs = mne.Epochs(raw, events, event_id=event_ids, on_missing='warn', preload=True)
    features = epochs.get_data()
    labels = epochs.events[:, -1]
    all_features.append(features)
    all_labels.append(labels)

features = np.concatenate(all_features, axis=0)
labels = np.concatenate(all_labels, axis=0)

scaler = StandardScaler()
features = scaler.fit_transform(features.reshape(features.shape[0], -1)).reshape(features.shape)
features = features[..., np.newaxis]

labels = to_categorical(labels - event_ids[0])

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")

# Build Inception-like Model
def build_inception(input_shape, num_classes):
    def inception_module(x, filters):
        conv1x1 = Conv2D(filters, (1, 1), activation='relu', padding='same')(x)
        conv1x1 = BatchNormalization()(conv1x1)
        
        conv3x3 = Conv2D(filters, (1, 3), activation='relu', padding='same')(x)
        conv3x3 = BatchNormalization()(conv3x3)
        
        conv5x5 = Conv2D(filters, (1, 5), activation='relu', padding='same')(x)
        conv5x5 = BatchNormalization()(conv5x5)
        
        pool_proj = MaxPooling2D((1, 3), strides=(1, 1), padding='same')(x)
        pool_proj = Conv2D(filters, (1, 1), activation='relu', padding='same')(pool_proj)
        pool_proj = BatchNormalization()(pool_proj)
        
        return concatenate([conv1x1, conv3x3, conv5x5, pool_proj], axis=-1)

    inputs = Input(shape=input_shape)
    x = Conv2D(32, (1, 7), activation='relu', padding='same')(inputs)
    x = BatchNormalization()(x)
    x = MaxPooling2D((1, 3))(x)
    x = inception_module(x, 32)
    x = inception_module(x, 64)
    x = Flatten()(x)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs, outputs)
    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

input_shape = (features.shape[1], features.shape[2], 1)
num_classes = labels.shape[1]

inception_model = build_inception(input_shape, num_classes)
inception_model.summary()

callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=1)
]

history = inception_model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=50,
    batch_size=32,
    callbacks=callbacks,
    verbose=1
)

test_loss, test_accuracy = inception_model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

y_pred_probs = inception_model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)
y_test_classes = np.argmax(y_test, axis=1)

print("\nClassification Report:")
print(classification_report(y_test_classes, y_pred))

conf_matrix = confusion_matrix(y_test_classes, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=range(num_classes), yticklabels=range(num_classes))
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

# ROC-AUC Calculation and Visualization
y_test_binarized = label_binarize(y_test_classes, classes=range(num_classes))

fpr = {}
tpr = {}
roc_auc = {}

# Compute ROC and AUC for each class
for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], y_pred_probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Print ROC-AUC for each class
print("\nROC-AUC for each class:")
for i in range(num_classes):
    print(f"Class {i}: AUC = {roc_auc[i]:.4f}")

# Compute the average ROC-AUC across all classes (macro-average)
average_roc_auc = np.mean(list(roc_auc.values()))
print(f"\nAverage ROC-AUC: {average_roc_auc:.4f}")

# Plot ROC Curves
plt.figure(figsize=(10, 8))
for i in range(num_classes):
    plt.plot(fpr[i], tpr[i], label=f"Class {i} (AUC = {roc_auc[i]:.2f})")

plt.plot([0, 1], [0, 1], color="navy", linestyle="--")
plt.title("Multi-Class ROC Curves")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
plt.show()

# Training History
plt.figure()
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

plt.figure()
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_curve, auc, classification_report, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt

# Function to build the model
def build_inception(input_shape, num_classes):
    def inception_module(x, filters):
        conv1x1 = Conv2D(filters, (1, 1), activation='relu', padding='same')(x)
        conv1x1 = BatchNormalization()(conv1x1)

        conv3x3 = Conv2D(filters, (1, 3), activation='relu', padding='same')(x)
        conv3x3 = BatchNormalization()(conv3x3)

        conv5x5 = Conv2D(filters, (1, 5), activation='relu', padding='same')(x)
        conv5x5 = BatchNormalization()(conv5x5)

        pool_proj = MaxPooling2D((1, 3), strides=(1, 1), padding='same')(x)
        pool_proj = Conv2D(filters, (1, 1), activation='relu', padding='same')(pool_proj)
        pool_proj = BatchNormalization()(pool_proj)

        return concatenate([conv1x1, conv3x3, conv5x5, pool_proj], axis=-1)

    inputs = Input(shape=input_shape)
    x = Conv2D(32, (1, 7), activation='relu', padding='same')(inputs)
    x = BatchNormalization()(x)
    x = MaxPooling2D((1, 3))(x)
    x = inception_module(x, 32)
    x = inception_module(x, 64)
    x = Flatten()(x)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs, outputs)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Prepare data for StratifiedKFold
labels_classes = np.argmax(labels, axis=1)  # Convert one-hot to class indices
num_classes = labels.shape[1]
input_shape = (features.shape[1], features.shape[2], 1)

# StratifiedKFold setup
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

fold = 1
accuracies = []
losses = []
roc_aucs = []

for train_idx, test_idx in skf.split(features, labels_classes):
    print(f"\nTraining Fold {fold}...")

    # Split the data
    X_train_cv, X_test_cv = features[train_idx], features[test_idx]
    y_train_cv, y_test_cv = labels[train_idx], labels[test_idx]

    # Build the model
    model = build_inception(input_shape, num_classes)

    # Train the model
    history = model.fit(X_train_cv, y_train_cv, epochs=10, batch_size=32, verbose=0)

    # Evaluate accuracy and loss
    loss, accuracy = model.evaluate(X_test_cv, y_test_cv, verbose=0)
    losses.append(loss)
    accuracies.append(accuracy)
    print(f"Fold {fold} Accuracy: {accuracy:.4f}")
    print(f"Fold {fold} Loss: {loss:.4f}")

    # Predict probabilities
    y_pred_probs = model.predict(X_test_cv)

    # Compute ROC-AUC for each class
    fpr, tpr, roc_auc = {}, {}, {}
    for i in range(num_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_cv[:, i], y_pred_probs[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Average ROC-AUC for the fold
    avg_roc_auc = np.mean(list(roc_auc.values()))
    roc_aucs.append(avg_roc_auc)
    print(f"Fold {fold} Average ROC-AUC: {avg_roc_auc:.4f}")

    # Plot ROC curves for the fold
    plt.figure(figsize=(10, 8))
    for i in range(num_classes):
        plt.plot(fpr[i], tpr[i], label=f"Class {i} (AUC = {roc_auc[i]:.2f})")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.title(f"ROC Curve for Fold {fold}")
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc='lower right')
    plt.grid()
    plt.show()

    fold += 1

# Final Cross-Validation Results
print("\nCross-Validation Results:")
print(f"Average Accuracy: {np.mean(accuracies) * 100:.2f}%")
print(f"Standard Deviation of Accuracy: {np.std(accuracies) * 100:.2f}%")
print(f"Average Loss: {np.mean(losses):.4f}")
print(f"Standard Deviation of Loss: {np.std(losses):.4f}")
print(f"Average ROC-AUC: {np.mean(roc_aucs):.4f}")
print(f"Standard Deviation of ROC-AUC: {np.std(roc_aucs):.4f}")
```
